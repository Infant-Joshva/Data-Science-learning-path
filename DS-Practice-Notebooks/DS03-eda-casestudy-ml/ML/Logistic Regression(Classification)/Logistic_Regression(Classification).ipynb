## Structured Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
X=pd.DataFrame(data.data,columns=data.feature_names)
X
```

```python
Y=pd.DataFrame(data.target,columns=['Target'])
Y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Lg=LogisticRegression()
```

```python
Lg.fit(X_train,Y_train)
```

```python
Lg.intercept_,Lg.coef_
```

```python
Lg.predict(X_test)
```

```python
Y_test
```

```python
Y_Train_pred=Lg.predict(X_train)
Y_Test_pred=Lg.predict(X_test)
```

```python
Y['Target'].value_counts()
```

```python
Y_Train_pred,Y_Test_pred
```

```python
Y_train,Y_test
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report
```

```python
print('Train Accuracy',accuracy_score(Y_train,Y_Train_pred))
print('Test Accuracy',accuracy_score(Y_test,Y_Test_pred))
print('--'*50)
print('Train Precision',precision_score(Y_train,Y_Train_pred))
print('Test Precision',precision_score(Y_test,Y_Test_pred))
print('--'*50)
print('Train Recall',recall_score(Y_train,Y_Train_pred))
print('Test Recall',recall_score(Y_test,Y_Test_pred))
print('--'*50)
print('Train F1 Score',f1_score(Y_train,Y_Train_pred))
print('Test F1 Score',f1_score(Y_test,Y_Test_pred))
print('--'*50)
```

```python
print('Confusion Matrix')
print('Train',confusion_matrix(Y_train,Y_Train_pred))
print('Test',confusion_matrix(Y_test,Y_Test_pred))
print('--'*50)
print('Classification Report')
print('Train',classification_report(Y_train,Y_Train_pred))
print('Test',classification_report(Y_test,Y_Test_pred))
print('--'*50)
```

```python
yes_data={
    'Features':[],
    'Target':[]

}

for i in os.listdir('/content/brain_tumor_dataset/yes'):
  path=os.path.join('/content/brain_tumor_dataset/yes',i)
  img=cv2.imread(path)
  resize_img=cv2.resize(img,(100,100))
  flatten_img=resize_img.flatten()
  yes_data['Features'].append(flatten_img)
  yes_data['Target'].append(1)
```

```python
no_data={
    'Features':[],
    'Target':[]

}

for i in os.listdir('/content/brain_tumor_dataset/no'):
  path=os.path.join('/content/brain_tumor_dataset/no',i)
  img=cv2.imread(path)
  resize_img=cv2.resize(img,(100,100))
  flatten_img=resize_img.flatten()
  no_data['Features'].append(flatten_img)
  no_data['Target'].append(0)
```

```python
no_data
```

```python
yes_data
```

```python
df_yes=pd.DataFrame(yes_data)
df_no=pd.DataFrame(no_data)
```

```python
df=pd.concat([df_yes,df_no])
df
```

```python
X=df.drop('Target',axis=1)
Y=df['Target']
```

```python
X.iloc[0]
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
X_train
```

```python
X_train_array=np.array(X_train['Features'].tolist())
X_test_array=np.array(X_test['Features'].tolist())
```

```python
len(X_train_array[0])
```

```python
Lg=LogisticRegression()
```

```python
Lg.fit(X_train_array,Y_train)
```

```python
Y_tran_pred=Lg.predict(X_train_array)
Y_test_pred=Lg.predict(X_test_array)
```

```python
print('Train Report',classification_report(Y_train,Y_tran_pred))
print('Test Report',classification_report(Y_test,Y_test_pred))
```
