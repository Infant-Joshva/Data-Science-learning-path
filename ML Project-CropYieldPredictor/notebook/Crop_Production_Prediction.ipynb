#### Pivoting



```
# Pivoting the data to get structure look
```



````python
df
````

````python
# Create a combined pivot column
df["Element_Unit"] = df["Element"] + "_" + df["Unit"].str.replace("/", "_")
````

````python
df["Element_Unit"]
````

````python
df.head(10)
````

````python
#Pivot

pivot_df = df.pivot_table(
    index=["Area", "Item", "Year"],
    columns="Element_Unit",
    values="Value",
    aggfunc="first"
    ).reset_index()

pivot_df
````

````python
pivot_df.head(10)
````

````python
pivot_df.columns
````

````python
pivot_df = pivot_df.drop(
    pivot_df.filter(regex="An").columns,
    axis=1
)
````

````python
pivot_df.columns
````

````python
pivot_df.drop(['Production_1000 No'],axis=1,inplace=True)
````

````python
# 'Stocks_No'

pivot_df.drop(['Stocks_No'],axis=1,inplace=True)
````

````python
pivot_df.reset_index(drop=True, inplace=True)
````
