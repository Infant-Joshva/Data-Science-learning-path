###### Doing Hypothesis Test for require column

```python
#Hypothesis test

for col in num_con_df.columns:

  # class A and B scores
  class_a = kidney_df[kidney_df['Status']=='ckd'][col]
  class_b = kidney_df[kidney_df['Status']=='notckd'][col]

  # Perform two-sample t-test (assume equal variances)
  t_statistic, p_value = stats.ttest_ind(class_a, class_b)

  # Print the results
  print("T-statistic:", t_statistic)
  print("P-value:", p_value)

  # Interpretation
  alpha = 0.05  # significance level
  if p_value < alpha:
    print(f"Reject the null hypothesis. There is a significant difference in {col} between Disease_Status (Yes) and Disease_Status (No).")
    print('-'*150)
  else:
    print(f"Fail to reject the null hypothesis. There is no significant difference in {col} between Disease_Status (Yes) and Disease_Status (No).")
    print('-'*150)
```

```python
X = kidney_df.drop('Status', axis=1)
Y = kidney_df['Status']
```

```python
# Seperating Numerical and Categorical Features
# For 1st we seperating Numerical into 2 types one is Continues and Discreat

num_feature=X[X.select_dtypes(include=['int64','float64','int32']).columns]
num_feature
```

```python
Y
```

```python
# Seperating Numerical and Categorical Features
# For 1st we seperating Numerical into 2 types one is Continues and Discreat

num_feature_con = num_feature[['age', 'bp', 'sg', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wc', 'rc']].reset_index(drop=True) # Con Values
num_featurs_dis = num_feature[['al','su']].reset_index(drop=True) # Dis Values
```

```python
num_featurs_dis.columns
```

```python
cat_features = X[X.select_dtypes(include='object').columns].reset_index(drop=True)
cat_features
```

```python
cat_features_TR = cat_features[['rbc','pc','pcc','ba','htn','dm','cad','appet','pe','ane']]
cat_features_TR
```

```python
cat_features_TR_cols = cat_features_TR.columns
cat_features_TR_cols
```

```python
# Finding Cross Tab to get order for Cat column - Target Encoding

# Loop through each categorical column

# This will store the results
cat_feature_TR_cols_cross_list = []

for col in cat_features_TR_cols:
    crosstab = pd.crosstab(kidney_df[col], kidney_df['Status'])

    # Compute the attrition ratio
    crosstab['Status_Ratio'] = crosstab['ckd'] / (crosstab['ckd'] + crosstab['notckd'])

    # Append the sorted index (category names) by attrition ratio
    sorted_index = crosstab.sort_values(by='Status_Ratio', ascending=False).index
    cat_feature_TR_cols_cross_list.append(sorted_index)

cat_feature_TR_cols_cross_list
```

```python
cat_feature_TR_cols_cross_list
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

##### Encoding-Pipeline

```python
# Creating transformer for Numerical con columns

num_con_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

num_con_transformer
```

```python
# Creating transformer for Numerical dis columns

num_dis_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('scaler', StandardScaler())
])

num_dis_transformer
```

```python
# Creating transformer for cat odinal column

cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OrdinalEncoder(categories=[['abnormal', 'normal'],
     ['abnormal', 'normal'],
     ['present', 'notpresent'],
     ['present', 'notpresent'],
     ['yes', 'no'],
     ['yes', 'no'],
     ['yes', 'no'],
     ['poor', 'good'],
     ['yes', 'no'],
     ['yes', 'no']]))
])

cat_transformer
```

```python
# Combining the all transformer as a full loaded pipeline with column transformer

preprocessor = ColumnTransformer(
    transformers=[
        ('num_con', num_con_transformer, num_feature_con.columns),
        ('num_dis', num_dis_transformer, num_featurs_dis.columns),
        ('cat', cat_transformer, cat_features_TR.columns)
    ])
preprocessor
```

##### LogReg

```python
# Creating pipeline for LogReg

logreg_pipeline = IMBPipeline(steps=[
    ('preprocessor', preprocessor),
    ('Sampling',SMOTE(random_state=42)),
    ('classifier', LogisticRegression(random_state=42))
])

logreg_pipeline.fit(X_train, Y_train)

Y_train_pred=logreg_pipeline.predict(X_train)
Y_test_pred=logreg_pipeline.predict(X_test)

print('Confusion Matrix Train')
print('--'*50)
print(confusion_matrix(Y_train,Y_train_pred))
print('Confusion Matrix Test')
print('--'*50)
print(confusion_matrix(Y_test,Y_test_pred))
print('**'*50)
print('Classification Report Train')
print('--'*50)
print(classification_report(Y_train,Y_train_pred))
print('Classification Report Test')
print('--'*50)
print(classification_report(Y_test,Y_test_pred))
print('--'*50)
```

##### Random Forest

```python
# Creating pipeline for Random Forest

RF_pipeline = IMBPipeline(steps=[
    ('preprocessor', preprocessor),
    ('Sampling',SMOTE(random_state=42)),
    ('classifier', RandomForestClassifier(random_state=42))
])

RF_pipeline.fit(X_train, Y_train)

Y_train_pred=RF_pipeline.predict(X_train)
Y_test_pred=RF_pipeline.predict(X_test)

print('Confusion Matrix Train')
print('--'*50)
print(confusion_matrix(Y_train,Y_train_pred))
print('Confusion Matrix Test')
print('--'*50)
print(confusion_matrix(Y_test,Y_test_pred))
print('**'*50)
print('Classification Report Train')
print('--'*50)
print(classification_report(Y_train,Y_train_pred))
print('Classification Report Test')
print('--'*50)
print(classification_report(Y_test,Y_test_pred))
print('--'*50)
```
