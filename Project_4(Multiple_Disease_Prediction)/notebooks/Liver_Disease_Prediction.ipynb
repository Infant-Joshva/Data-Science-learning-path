##### Doing Hypothesis Test for require column


**Need to confirm whether these column is required or not with hypothesis test**
*   Total_Protiens
*   Albumin
*   Albumin and Globulin



```python
# Total_Protiens

# class A and B scores
class_A = liver_df[liver_df['Disease_Status'] == 'Yes']['Total_Protiens']
class_B = liver_df[liver_df['Disease_Status'] == 'No']['Total_Protiens']

# Perform two-sample t-test (assume equal variances)
t_statistic, p_value = ttest_ind(class_A, class_B)

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in Total_Protiens between Disease_Status (Yes) and Disease_Status (No).")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in Total_Protiens between Disease_Status (Yes) and Disease_Status (No).")
```

```python
# Albumin

# class A and B scores
class_A = liver_df[liver_df['Disease_Status'] == 'Yes']['Albumin']
class_B = liver_df[liver_df['Disease_Status'] == 'No']['Albumin']

# Perform two-sample t-test (assume equal variances)
t_statistic, p_value = ttest_ind(class_A, class_B)

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in Albumin between Disease_Status (Yes) and Disease_Status (No).")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in Albumin between Disease_Status (Yes) and Disease_Status (No).")
```

```python
# Albumin_and_Globulin_Ratio

# class A and B scores
class_A = liver_df[liver_df['Disease_Status'] == 'Yes']['Albumin_and_Globulin_Ratio']
class_B = liver_df[liver_df['Disease_Status'] == 'No']['Albumin_and_Globulin_Ratio']

# Perform two-sample t-test (assume equal variances)
t_statistic, p_value = ttest_ind(class_A, class_B)

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in Albumin_and_Globulin_Ratio between Disease_Status (Yes) and Disease_Status (No).")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in Albumin_and_Globulin_Ratio between Disease_Status (Yes) and Disease_Status (No).")
```

### Train, Test Split and Scaling, Encoding

```python
# liver_df['Disease_Status_binary'] = liver_df['Disease_Status'].map({'Yes': 1, 'No': 0})
```

```python
X = liver_df.drop(['Disease_Status', 'Disease_Status_binary','Dataset'], axis=1)
Y = liver_df['Disease_Status']
```

```python
# Seperating Numerical and Categorical Features
# For 1st we seperating Numerical into 2 types one is Continues and Discreat

num_cols = X.select_dtypes(include=['int64', 'float64']).columns
cat_cols = X.select_dtypes(include=['object', 'category']).columns
```

```python
num_cols
```

##### Encoding-Pipeline

```python
# Creating transformer for Numerical con columns

num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

num_transformer
```

```python
# Finding Cross Tab to get order for Cat column - Target Encoding

crosstab = pd.crosstab(liver_df['Gender'], liver_df['Disease_Status'])
crosstab['Status_ratio'] = crosstab['Yes'] / (crosstab['Yes'] + crosstab['No'])
sorted_index = crosstab.sort_values(by='Status_ratio', ascending=False).index
sorted_index
```

```python
# Creating transformer for cat odinal column

cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OrdinalEncoder(categories=[sorted_index]))
])

cat_transformer
```

```python
# Combining the all transformer as a full loaded pipeline with column transformer

preprocess = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_cols),
        ('cat', cat_transformer, cat_cols)
    ])

preprocess
```

#### ML - Pipeline

##### Data Balancing

```python
from imblearn.pipeline import Pipeline as IMBPipeline
```

* We can choose ADASYN

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

##### LogReg

```python
# Creating pipeline for LogReg

logreg_pipeline = IMBPipeline(steps=[
    ('preprocess', preprocess),
    ('Sampling',ADASYN(random_state=42)),
    ('Poly',PolynomialFeatures(degree=3)),
    ('classifier', LogisticRegression(class_weight="balanced",penalty='l2', max_iter=300,multi_class='deprecated',intercept_scaling=1))
])

logreg_pipeline.fit(X_train,Y_train)

Y_train_pred=logreg_pipeline.predict(X_train)
Y_test_pred=logreg_pipeline.predict(X_test)

print('Confusion Matrix Train')
print('--'*50)
print(confusion_matrix(Y_train,Y_train_pred))
print('Confusion Matrix Test')
print('--'*50)
print(confusion_matrix(Y_test,Y_test_pred))
print('**'*50)
print('Classification Report Train')
print('--'*50)
print(classification_report(Y_train,Y_train_pred))
print('Classification Report Test')
print('--'*50)
print(classification_report(Y_test,Y_test_pred))
print('--'*50)
```
